```python
from sklearn.tree import export_graphviz
import pydotplus
from IPython.display import Imageimport pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import pydotplus
from IPython.display import Image
from io import StringIO

# Load dataset with original headers
df = pd.read_csv("diabetes.csv")
print(df.head())

# Define feature columns and target
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[feature_cols]
y = df['Outcome']

# Split data (70% train, 30% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Initialize Decision Tree Classifier (with entropy criterion and max depth 3)
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=1)

# Train classifier
clf.fit(X_train, y_train)

# Predict on test set
y_pred = clf.predict(X_test)

# Evaluation
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", metrics.confusion_matrix(y_test, y_pred))

# Visualize with matplotlib plot_tree
plt.figure(figsize=(20,12))
plot_tree(clf, 
          feature_names=feature_cols, 
          class_names=['No Diabetes', 'Diabetes'], 
          filled=True, 
          rounded=True,
          fontsize=12)
plt.show()

# Export tree to DOT format and create PNG image
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data, 
                feature_names=feature_cols, 
                class_names=['No Diabetes', 'Diabetes'],
                filled=True, rounded=True, special_characters=True)

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png("diabetes.png")

# Display PNG inline (Jupyter / IPython)
Image(filename="diabetes.png")

from io import StringIO

# Create a StringIO object to hold the dot data
dot_data = StringIO()

# Export the decision tree to DOT format and write to dot_data
export_graphviz(clf, 
                out_file=dot_data, 
                feature_names=feature_cols, 
                class_names=['No Diabetes', 'Diabetes'], 
                filled=True, 
                rounded=True, 
                special_characters=True)

# Create a graph from dot_data
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

# Save the graph as a PNG file
graph.write_png("diabetes.png")

# Display the image inline (Jupyter Notebook / IPython)
Image(filename="diabetes.png")
```

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier, export_graphviz, plot_tree
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import pydotplus
from IPython.display import Image
from io import StringIO

# Load dataset
df = pd.read_csv("diabetes.csv")

# Define features and target
feature_cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 
                'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']
X = df[feature_cols]
y = df['Outcome']

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

# Initialize Decision Tree
clf = DecisionTreeClassifier(criterion="entropy", max_depth=3, random_state=1)

# Train model
clf.fit(X_train, y_train)

# Predict
y_pred = clf.predict(X_test)

# Evaluate
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
print("Confusion Matrix:\n", metrics.confusion_matrix(y_test, y_pred))

# Plot tree using matplotlib
plt.figure(figsize=(20,12))
plot_tree(clf, 
          feature_names=feature_cols, 
          class_names=['No Diabetes', 'Diabetes'], 
          filled=True, rounded=True,
          fontsize=12)
plt.show()

# Export tree to dot format
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,
                feature_names=feature_cols,
                class_names=['No Diabetes', 'Diabetes'],
                filled=True, rounded=True,
                special_characters=True)

# Create graph from dot data
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())

# Save graph as PNG
graph.write_png("diabetes.png")

# Display PNG image inline (in Jupyter notebook or IPython)
Image(filename="diabetes.png")
```
